```python
import numpy as np
import matplotlib.pyplot as plt

print("Solution for Q1")
M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 100)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'r-', label=f'y = {M}x')
plt.plot(x, y2, 'b--', label=f'y = {M}sin(x)')
plt.grid(True)
plt.legend()
plt.title(f'Mathematical Functions with M = {M}')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print("Solution for Q2")
subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']
scores = [85, 92, 78, 88, 95]
df = pd.DataFrame({'Subjects': subjects, 'Scores': scores})

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Subjects', y='Scores', hue='Subjects', legend=False)
for i, v in enumerate(scores):
    plt.text(i, v + 1, str(v), ha='center')
plt.title('Subject Scores')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("Solution for Q3")
roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)
data = np.random.randn(50)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle('Data Visualization')

axes[0].plot(np.cumsum(data), 'g-')
axes[0].set_title('Cumulative Sum')
axes[0].grid(True)

noise = np.random.normal(0, 0.1, len(data))
axes[1].scatter(range(len(data)), data + noise, alpha=0.6)
axes[1].set_title('Scatter Plot with Noise')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print("Solution for Q4")
data = pd.read_csv('company_sales_data.csv')

plt.figure(figsize=(12, 6))
plt.plot(data['total_profit'], marker='o')
plt.title('Total Profit of All Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in products:
    plt.plot(data[product], label=product, marker='o')
plt.title('Product Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(15, 6))
product_sales = data[products].sum()
product_sales.plot(kind='bar')
plt.title('Total Sales by Product')
plt.xlabel('Products')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
